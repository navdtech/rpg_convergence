```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
S = 10
A = 10
y = 0.9
a = 1
```

```python
def hol(p):
  return 1/(1-1/p)

def fsa(p):

  S = 10
  A = 10
  y = 0.9
  a = 1
  return (2*(1+S*A)*A*y)/((1-y)**3)

def fs(p):
  q = hol(p)
  S = 10
  A = 10
  y = 0.9
  a = 1
  return (2*A*y)/((1-y)**3) + S*((2*A*y)/((1-y)**3))+ ((q-1)*3*A**(2-1/p))/((1-y)**2)

def f(p):
  q = hol(p)
  S = 10
  A = 10
  y = 0.9
  a = 1
  return  ((S*A)**(2-1/p))*(((q-1)*(A**2))/((1-y)**4)) +   ((2*A*y)/((1-y)**3))*((S*A)**(1/q)+1)
```

```python
fsa(2),fs(2),fs(2)
```

```python
A =[]
for p in np.arange(1.1,200,1):
  A.append([fsa(p),fs(p),f(p)])

```

```python
A = np.transpose(A)
```

```python

plt.plot(A[1])
plt.title('S-Rectangular Lp Robust Return')
plt.xlabel('p')
plt.ylabel('Smoothness constant')
plt.savefig('sp_smoothness')
```
