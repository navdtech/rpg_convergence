```python
def hol(p):
  return 1/(1-1/p)

def fnr():

  S = 10
  A = 10
  y = 0.9
  a = 1

  return (2*A*y)/((1-y)**3)

def fsa(p):

  S = 10
  A = 10
  y = 0.9
  a = 1
  return (2*(1+a)*A*y)/((1-y)**3)

def fs(p):
  q = hol(p)
  S = 10
  A = 10
  y = 0.9
  a = 1

  h = 1/(1-y)
  return 8*y*q*(1+a)*(A**(2*q))*(h**3)

def f(p):
  q = hol(p)
  S = 10
  A = 10
  y = 0.9
  a = 1
  return  ((S*A)**(2-1/p))*(((q-1)*(A**2))/((1-y)**4)) +   ((2*A*y)/((1-y)**3))*((S*A)**(1/q)+1)
```

```python
p = 100
fnr(), fsa(p)//fsa(p),fs(p)/fsa(p),f(p)/fsa(p)
```

```python
A =[]
for p in np.arange(1.1,200,1):
  A.append([fsa(p),fs(p),f(p)])

```

```python
A = np.transpose(A)
```

```python

plt.plot(A[1])
plt.title('S-Rectangular Lp Robust Return')
plt.xlabel('x = p-1.1')
plt.ylabel('Smoothness constant')
plt.savefig('sp_smoothness')
```
